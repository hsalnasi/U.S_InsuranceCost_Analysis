*building a helper function to help us edit any missing values or errors in our data*

```python
# create a function to handle any errors in capitalization
def edit_naming(dataframe): 
    edited_columns = []
    for column in dataframe.columns:
        column = column.title()
        edited_columns.append(column)
        
    dataframe.columns = edited_columns
        
    return dataframe
        
# create a function to round any decimal values to be float64 .
def round_column(dataframe, column,digit):
    rounded_lst = []
    # round the decimals
    for num in dataframe[column]:
        if isinstance(digit,int) and digit >= 0:
            num = round(num,digit)
           
        elif digit == 'None':
            num = round(num)
            
        rounded_lst.append(num)
    
    dataframe[column] = rounded_lst
    
    return dataframe

# add a symbol to the desiered column
def add_symbol(symbol,dataframe,column): 
    dataframe[column] = symbol + dataframe[column].astype(str)
    
    return dataframe
        
```

### *Analysis of the relationship between number of Children of individuals and Insurance costs*
#### in this section all data are read from the csv file and made into a list of dictionaries , then individuals are divided based on the number of chhildren they have and average insurance cost of each group is calculated and printed
#### at the end a dictionary containing the number of people in each category and their average insurance cost is saved in a dictionary called child_data

```python

with open('C:/Users/pouria zarrin/Desktop/snow-rep-US-insurancce/insurance.csv' , newline='') as insurance_data:
    data = list(csv.DictReader(insurance_data))

def child_analysis():
    #in this section categories are created based on the number of children
    #and based on these categories number of individuals and their costs are saved in children_count and children_cost respectively
    children_count = {}
    children_cost = {}
    for individual in data:
        x = individual['children']
        if x in children_count :
            children_count[x] += 1
        else :
            children_count[x] = 1
        if x in children_cost :
            children_cost[x] += round(float(individual['charges']))
        else :
            children_cost[x] = round(float(individual['charges']))


    #in this section keys and values of the children_cost dictionary are made into lists along side with the number of individuals in different variables
    #a loop is created that goes through the list of children_cost keys meaning it goes through each category of individuals and calculates the average 
    #cost of each category's insurance cost
    child_index_cost_value = list(children_cost.values())
    child_index_cost_keys = list(children_cost.keys())
    child_index_count = list(children_count.values())
    count = 0
    for individual in child_index_cost_keys:
        children_cost[individual] = round(float(child_index_cost_value[count] / child_index_count[count]))
        count += 1

    #finally a list is created that each catrgory is saved into a dictionary and appended to that list called child_data
    child_index_cost_value = list(children_cost.values())
    child_data = []
    for i in range(len(child_index_cost_keys)):
        print ('there are' , child_index_count[i] , 'people with' , child_index_cost_keys[i] , 'number of children that have' , child_index_cost_value[i] , 'average of insurance expenses')
        child_data.append({'children number category':int(child_index_cost_keys[i]) , 'individuals':child_index_count[i] , 'average cost':child_index_cost_value[i]})
    print ('\n',child_data)
child_analysis()
```
